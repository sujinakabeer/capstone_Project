```python
# Mapping month abbreviations to numeric values
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'June': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}


data['Month_Num'] = data['Month'].map(month_mapping)

print(data[['Month', 'Month_Num']].head())
```

```python
# Lineplot: Trend Analysis Over Months
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Month', y='Revenue', estimator='mean', errorbar=None)
plt.title('Trend Analysis Over Months')
plt.xlabel('Month')
plt.ylabel('Average Revenue')
plt.xticks(range(1, 13)) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Administrative_Duration', y='Revenue',errorbar=None)
plt.title('Revenue vs. Administrative Duration')
plt.xlabel('Administrative Duration')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='ProductRelated_Duration', y='Revenue', errorbar=None)
plt.title('Revenue vs. Product-Related Duration')
plt.xlabel('Product-Related Duration')
plt.ylabel('Revenue')
plt.show()
```

```python
# Barplot: Revenue by Visitor Type
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='VisitorType', y='Revenue', errorbar=None)
plt.title('Revenue by Visitor Type')
plt.xlabel('Visitor Type')
plt.ylabel('Average Revenue')
plt.show()
```

```python
# Scatterplot: Bounce Rates vs. Exit Rates
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='BounceRates', y='ExitRates', hue='Revenue', alpha=0.6)
plt.title('Bounce Rates vs. Exit Rates')
plt.xlabel('Bounce Rates')
plt.ylabel('Exit Rates')
plt.legend(title='Revenue')
plt.show()
```

```python
# Correlation Heatmap: Numerical Columns Only
plt.figure(figsize=(12, 8))
corr = data.select_dtypes(include=['float64', 'int64']).corr()  # Select numerical columns
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].median())
```

```python
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    if data[col].mode().empty:
        data[col].fillna("Unknown", inplace=True)
    else:
       data[col] = data[col].fillna(data[col].mode()[0])
```

```python
le = LabelEncoder()
for col in categorical_columns:
    data[col] = le.fit_transform(data[col])
```

```python
scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python
# Check distribution of target variable
sns.countplot(x='Revenue', data=data)
plt.title('Target Variable Distribution')
plt.show()
```

```python
# Create boxplots for numerical features
numerical_columns = data.select_dtypes(include=[np.number]).columns
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```
